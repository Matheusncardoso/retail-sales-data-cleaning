```python
# 1. Importa√ß√£o de bibliotecas
import pandas as pd
import numpy as np
from IPython.display import display
```

```python
# 2. Carregamento do Dataset
df = pd.read_csv('retail_sales_dirty.csv', sep=',')  # << aqui est√° o segredo
# Visualizar primeiras linhas
display(df.head())
```

```python
# Informa√ß√µes gerais sobre o dataset
display(df.info())
```

```python
# Verificar se existem valores nulos
display(df.isnull().sum())
```

```python
# Estat√≠sticas descritivas
display(df.describe(include='all'))
```

```python
# Remover linhas com valores nulos em 'Item', 'Price Per Unit' e 'Quantity'
df = df.dropna(subset=['Item', 'Price Per Unit', 'Quantity'])
```

```python
# Conferir se ainda existem valores nulos nessas colunas
display(df[['Item', 'Price Per Unit', 'Quantity']].isnull().sum())
```

```python
# Preencher valores nulos na coluna 'Discount Applied' como False
df['Discount Applied'] = df['Discount Applied'].fillna(False)
# Conferir novamente
display(df['Discount Applied'].isnull().sum())
```

```python
# Converter 'Transaction Date' para formato datetime
df.loc[:, 'Transaction Date'] = pd.to_datetime(df['Transaction Date'], errors='coerce')
# Conferir se houve convers√£o
df['Transaction Date'].dtypes
```

```python
# Criar uma nova coluna tempor√°ria para recalcular o total
df.loc[:, 'Total Spent Calculado'] = df['Price Per Unit'] * df['Quantity']
# Verificar se h√° diferen√ßa entre Total Spent registrado e o recalculado
(df['Total Spent'] - df['Total Spent Calculado']).abs().max()
```

```python
# 5. Exporta√ß√£o do Dataset Limpo
df = df.drop(columns=['Total Spent Calculado'])  # remove coluna tempor√°ria
df.to_csv('retail_sales_cleaned.csv', index=False)
print("‚úÖ Dataset exportado como 'retail_sales_cleaned.csv'")
```
